# 변수들에 대한 이상치 검사

시력 : 최대값이 9.9이므르 이상치처리 필요

공복 혈당 : 정상 수치가 70~100임, 126이상일 시 당뇨병으로 분류 - 담배와 관련 있음 -> 이상치 처리에 신중해야 함 : 이상치 처리를 안해도 될 거 같다.

혈압 : 혈압의 경우 이완기 수축기가 있는데 해당 데이터셋은 이완기 혈압으로 추측됨, 정상의 경우 80, 저혈압은 60이하, 고혈압은 90이상이라 함 - 담배와 관련 있음

중성지방 : 150미만 정상, 150 ~ 199 경계, 200 ~ 499 높음, 500 ~ 아주 높음 - 999가 있을 수 있는 수치만 75%와 비교했을 때 너무 높기 때문에 스케일링이 필요해 보임 - 담배와 관련 있음

혈청 크레아티닌 : 정상 범위 남성 0.7~1.4, 여성 0.5 ~ 1.1

헤모글로빈 : 남성 13.5-17.5 정상, 여성 12.5-15.5 정상 - 이상치 처리 필요 없어보임

시력의 이상치 검사

- 시력의 경우 흡연 여부와 큰 연관이 없을 것이라 판단해 이상치 처리를 중간값으로 할 예정

```python
#시력의 이상치 검사 - IQR(사분위수범위) : 적절하지 않은 것 같다
Q1 = train['시력'].quantile(0.25)
Q3 = train['시력'].quantile(0.75)
IQR = Q3 - Q1
outliers = train['시력'][(train['시력'] < (Q1 - 1.5 * IQR)) | (train['시력'] > (Q3 + 1.5 * IQR))]
print(outliers)
```

```python
# 시력의 이상치 검사 - z-score : 적절한 것 같다

시력_평균 = train['시력'].mean()
시력_표준편차 = train['시력'].std()
시력_중간값 = train['시력'].median()
threshold = 3

outlier = []

for i in train['시력']:
    z = (i-시력_평균)/시력_표준편차
    if z > threshold:
        outlier.append(i)

outlier
```

```python
#시력 이상치 처리 - 중간값으로

train['시력'] = train['시력'].apply(lambda x: 시력_중간값 if (x - 시력_평균) / 시력_표준편차 > threshold else x)

```

공복 혈당의 이상치 검사

```python
#공복 혈당의 이상치 검사 - IQR(사분위수범위) :
Q1 = train['공복 혈당'].quantile(0.25)
Q3 = train['공복 혈당'].quantile(0.75)
IQR = Q3 - Q1
outliers = train['공복 혈당'][(train['공복 혈당'] < (Q1 - 1.5 * IQR)) | (train['공복 혈당'] > (Q3 + 1.5 * IQR))]
print(outliers)
```

```python
#공복 혈당의 이상치 검사 - z-score

공복_혈당_평균 = train['공복 혈당'].mean()
공복_혈당_표준편차 = train['공복 혈당'].std()
공복_혈당_중간값 = train['공복 혈당'].median()
threshold = 3

outlier = []

for i in train['공복 혈당']:
    z = (i-공복_혈당_평균)/공복_혈당_표준편차
    if z > threshold:
        outlier.append(i)

outlier
```

```python
#혈압의 이상치 검사 - IQR(사분위수범위) : 적절하지 않은 것 같다
Q1 = train['혈압'].quantile(0.25)
Q3 = train['혈압'].quantile(0.75)
IQR = Q3 - Q1
outliers = train['혈압'][(train['혈압'] < (Q1 - 1.5 * IQR)) | (train['혈압'] > (Q3 + 1.5 * IQR))]
print(outliers)
```

```python
# 혈압의 이상치 검사 - z-score : 적절하지 않은 것 같음

혈압_평균 = train['혈압'].mean()
혈압_표준편차 = train['혈압'].std()
혈압_중간값 = train['혈압'].median()
threshold = 3

outlier = []

for i in train['혈압']:
    z = abs((i-혈압_평균)/혈압_표준편차 )
    if z > threshold:
        outlier.append(i)


outlier
```

혈압의 경우 극단적으로 높은 값보다 낮은 값의 이상치가 있어 LCL로 처리

```python
k = 3  # 일반적으로 3으로 설정

LCL = 혈압_평균 - k * 혈압_표준편차

print(f' LCL: {LCL}')
```

```python
# 혈압은 흡연과 관계가 유의미하다고 판단되어 LCL값으로 대체함
train['혈압'] = train['혈압'].apply(lambda x: LCL if ( x < LCL) else x)
```

```python
train['혈압'].min()
```

```python
#중성지방 사분위수 - 적절하지 않음

Q1 = train['중성 지방'].quantile(0.25)
Q3 = train['중성 지방'].quantile(0.75)
IQR = Q3 - Q1
outliers = train['중성 지방'][(train['중성 지방'] < (Q1 - 1.5 * IQR)) | (train['중성 지방'] > (Q3 + 1.5 * IQR))]
print(outliers.sort_values())
```

```python
# 중성 지방 z - score - 적절하지 않은 것 같음

중성_지방_평균 = train['중성 지방'].mean()
중성_지방_표준편차 = train['중성 지방'].std()
중성_지방_중간값 = train['중성 지방'].median()


outlier = []

for i in train['중성 지방']:
    z = abs((i-중성_지방_평균)/중성_지방_표준편차 )
    if z > threshold:
        outlier.append(i)


outlier
```

```python
# LCL, UCL 방식 모두 적절하지 않은 것 같다.

k = 3  # 일반적으로 3으로 설정
UCL = 중성_지방_평균 + k * 중성_지방_표준편차
LCL = abs(중성_지방_평균 - k * 중성_지방_표준편차)

print(f'UCL: {UCL}, LCL: {LCL}')
```

```python
train['중성 지방'] = train['중성 지방'].apply(lambda x: 600 if ( x > 600) else x)
```

```python
#혈청 크레아티닌 사분위수 - 적절하지 않음

Q1 = train['혈청 크레아티닌'].quantile(0.25)
Q3 = train['혈청 크레아티닌'].quantile(0.75)
IQR = Q3 - Q1

outliers = train['혈청 크레아티닌'][(train['혈청 크레아티닌'] < (Q1 - 1.5 * IQR)) | (train['혈청 크레아티닌'] > (Q3 + 1.5 * IQR))]
print(outliers.sort_values())
```

```python
# 혈청 크레아티닌 UCL

혈청_크레아티닌_평균 = train['혈청 크레아티닌'].mean()
혈청_크레아티닌_표준편차 = train['혈청 크레아티닌'].std()


k = 3  # 일반적으로 3으로 설정
UCL = 혈청_크레아티닌_평균 + k * 혈청_크레아티닌_표준편차
LCL = abs(혈청_크레아티닌_평균 - k * 혈청_크레아티닌_표준편차)

print(f'UCL: {UCL}, LCL: {LCL}')
```

```python
train['혈청 크레아티닌'] = train['혈청 크레아티닌'].apply(lambda x: UCL if ( x > UCL) else x)
```

```python
#고밀도지단백 이상치 처리
train['고밀도지단백'] = train['고밀도지단백'].apply(lambda x: UCL if ( x > UCL) else x)
```

```python
#저밀도지단백의 이상치 검사

저밀도지단백_평균 = train['저밀도지단백'].mean()
저밀도지단백_표준편차 = train['저밀도지단백'].std()
k = 3
UCL = 저밀도지단백_평균 + k * 저밀도지단백_표준편차
LCL = abs(저밀도지단백_평균 - k * 저밀도지단백_표준편차)

print(f'UCL: {UCL}, LCL: {LCL}')

outlier = []


for i in train['저밀도지단백']:
    z = abs((i-저밀도지단백_평균)/저밀도지단백_표준편차 )
    if z > threshold:
        outlier.append(i)


outlier
```
